```python
%sql

WITH monthly AS (
  SELECT
    DATE_TRUNC('month', order_date) AS month,
    category,
    SUM(revenue) AS revenue,
    SUM(units) AS units
  FROM orders
  GROUP BY 1,2
), month_overall AS (
  SELECT month, SUM(revenue) AS total_revenue
  FROM monthly
  GROUP BY 1
)
SELECT monthly.month, monthly.category, monthly.revenue,
       ROUND((m.revenue / AVG(m.revenue) OVER (PARTITION BY m.category))::numeric,2) AS seasonality_index,
       mo.total_revenue
FROM monthly m
JOIN month_overall mo USING (month)
WHERE m.category IN ('Outerwear')
ORDER BY m.month;
```
