### 1. Orders and Order Items Analysis

#### Revenue and Profitability
- Total revenue = SUM(price_usd) from orders or order_items.
- Cost of Goods Sold (COGS) = SUM(cogs_usd).
- Gross profit = Revenue  COGS.
- Profit margin per product or order.

```python
%%sql

SELECT 
    SUM(oi.price_usd) AS total_revenue,
    SUM(oi.cogs_usd) AS total_cogs,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd)) AS gross_profit,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd)) / (SUM(oi.price_usd)) AS profit_margin
FROM 
    order_items oi;
```

```python
%%sql
-- PROFITABILITY PER ORDER

WITH top_orders_cte AS (
SELECT 
    o.order_id,
    SUM(oi.price_usd) AS order_revenue,
    SUM(oi.cogs_usd) AS order_cogs,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd)) AS order_gross_profit,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd)) / (SUM(oi.price_usd)) AS order_profit_margin
FROM orders o
JOIN order_items oi
    ON o.order_id = oi.order_id  
GROUP BY o.order_id
ORDER BY o.order_id
)

SELECT * FROM top_orders_cte LIMIT 10;
```

```python
%%sql
-- PROFITABILITY PER DAY

WITH revenue_trend_cte AS (
    SELECT 
        DATE(o.created_at) AS order_date,
        DATE_FORMAT(o.created_at, '%Y-%m') AS order_month,
        DATE_FORMAT(o.created_at, '%Y') AS order_year,
        SUM(oi.price_usd) AS revenue,
        SUM(oi.cogs_usd) AS cogs,
        SUM(oi.price_usd) - SUM(oi.cogs_usd) AS gross_profit,
        (SUM(oi.price_usd) - SUM(oi.cogs_usd)) / SUM(oi.price_usd) AS profit_margin
    FROM orders o
    JOIN order_items oi
        ON o.order_id = oi.order_id
    GROUP BY DATE(o.created_at)
)

SELECT *
FROM revenue_trend_cte
LIMIT 10;
```

#### Product Returns/Refund Rate

- Refund rate = refunds  revenue
    - (High refund rate = risk; product may have quality issues or expectation mismatch)

- This ranks products from highest refund rate (worst performing) to lowest refund rate.

```python
%%sql

SELECT
    p.product_id,
    p.product_name,
    SUM(oi.price_usd) AS revenue,
    SUM(COALESCE(r.refund_amount_usd, 0)) AS refunds,
    (SUM(COALESCE(r.refund_amount_usd, 0)) / SUM(oi.price_usd)) AS refund_rate
FROM products p
JOIN order_items oi 
    ON p.product_id = oi.product_id
LEFT JOIN order_item_refunds r 
    ON oi.order_item_id = r.order_item_id
GROUP BY p.product_id, p.product_name
HAVING revenue > 0
ORDER BY refund_rate DESC;
```

### 3. Refund Analysis

```python
%%sql

WITH daily_profitability_cte AS (
    SELECT
        DATE(o.created_at) AS order_date,
        SUM(oi.price_usd) AS revenue,
        SUM(oi.cogs_usd) AS cogs,
        (SUM(oi.price_usd) - SUM(oi.cogs_usd)) AS gross_profit,
        COALESCE(SUM(r.refund_amount_usd), 0) AS refunds,
        -- Net Profit = Revenue - COGS - Refunds
        (SUM(oi.price_usd) - SUM(oi.cogs_usd) - COALESCE(SUM(r.refund_amount_usd), 0)) AS net_profit,
        -- Profit Margin = Net Profit / Revenue
        ROUND(
            (SUM(oi.price_usd) - SUM(oi.cogs_usd) - COALESCE(SUM(r.refund_amount_usd), 0))
            / NULLIF(SUM(oi.price_usd), 0),
            4
        ) AS net_profit_margin,
        -- Refund Rate = refunds / revenue
        ROUND(
            COALESCE(SUM(r.refund_amount_usd), 0) / NULLIF(SUM(oi.price_usd), 0),
            4
        ) AS refund_rate

    FROM orders o
    JOIN order_items oi 
        ON o.order_id = oi.order_id
    LEFT JOIN order_item_refunds r 
        ON oi.order_item_id = r.order_item_id

    GROUP BY DATE(o.created_at)
)

SELECT *
FROM daily_profitability_cte
ORDER BY order_date
LIMIT 10;
```

```python
%%sql

CREATE OR REPLACE VIEW order_analytics_view AS
SELECT
    /* -------------------- ORDER-LEVEL ----------------------- */
    o.order_id,
    o.created_at AS order_created_at,
    o.user_id AS user_id,
    o.website_session_id,
    o.primary_product_id,
    
    -- Order totals
    SUM(oi.price_usd) AS order_revenue,
    SUM(oi.cogs_usd) AS order_cogs,
    COALESCE(SUM(oif.refund_amount_usd), 0) AS order_refunds,
    
    -- Profits
    (SUM(oi.price_usd) - SUM(oi.cogs_usd)) AS order_gross_profit,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd) - COALESCE(SUM(oif.refund_amount_usd), 0)) AS order_net_profit,

    /* -------------------- PRODUCT-LEVEL ----------------------- */
    oi.product_id,
    p.product_name,
    SUM(oi.price_usd) AS product_revenue,
    SUM(oi.cogs_usd) AS product_cogs,
    COALESCE(SUM(oif.refund_amount_usd), 0) AS product_refunds,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd)) AS product_gross_profit,
    (SUM(oi.price_usd) - SUM(oi.cogs_usd) - COALESCE(SUM(oif.refund_amount_usd), 0)) AS product_net_profit,

    /* -------------------- USER-LEVEL ----------------------- */
    u.total_user_orders,
    u.total_user_revenue,
    u.first_order_date,
    u.last_order_date,

    /* -------------------- SESSION-LEVEL -------------------- */
    ws.is_repeat_session,
    ws.utm_source,
    ws.utm_campaign,
    ws.utm_content,
    ws.device_type,
    ws.http_referer,
    wps.pageviews_count AS session_pageviews

FROM orders o
JOIN order_items oi 
    ON o.order_id = oi.order_id
LEFT JOIN order_item_refunds oif
    ON oi.order_item_id = oif.order_item_id
JOIN products p
    ON oi.product_id = p.product_id
LEFT JOIN website_sessions ws
    ON o.website_session_id = ws.website_session_id

/* -------------------- USER METRICS SUBQUERY -------------------- */
LEFT JOIN (
    SELECT
        o2.user_id,
        COUNT(DISTINCT o2.order_id) AS total_user_orders,
        SUM(oi2.price_usd) AS total_user_revenue,
        MIN(o2.created_at) AS first_order_date,
        MAX(o2.created_at) AS last_order_date
    FROM orders o2
    JOIN order_items oi2 ON o2.order_id = oi2.order_id
    GROUP BY o2.user_id
) u ON o.user_id = u.user_id

/* -------------------- SESSION PAGEVIEW COUNTS -------------------- */
LEFT JOIN (
    SELECT 
        website_session_id,
        COUNT(*) AS pageviews_count
    FROM website_pageviews
    GROUP BY website_session_id
) wps
ON o.website_session_id = wps.website_session_id

GROUP BY 
    o.order_id,
    oi.product_id;
```

### 7. Referral Effectiveness

#### Revenue & Conversion by http_referer

```python
%%sql

SELECT
    ws.http_referer,
    COUNT(DISTINCT ws.website_session_id) AS total_sessions,
    COUNT(DISTINCT o.order_id) AS total_orders,
    ROUND(
        COUNT(DISTINCT o.order_id) / COUNT(DISTINCT ws.website_session_id),
        4
    ) AS conversion_rate,
    SUM(o.price_usd) AS total_revenue
FROM website_sessions ws
LEFT JOIN orders o
    ON ws.website_session_id = o.website_session_id
GROUP BY ws.http_referer
ORDER BY conversion_rate DESC;
```

### 8. Device and Traffic

#### a. Conversion Rate by device type

```python
%%sql

SELECT
    ws.device_type,
    COUNT(DISTINCT ws.website_session_id) AS total_sessions,COUNT(DISTINCT o.order_id) AS total_orders,
    ROUND(
        COUNT(DISTINCT o.order_id) / COUNT(DISTINCT ws.website_session_id),
        4
    ) AS conversion_rate
FROM website_sessions ws
LEFT JOIN orders o
    ON ws.website_session_id = o.website_session_id
GROUP BY ws.device_type
ORDER BY conversion_rate DESC;
```

#### b. Revenue by device type

```python
%%sql

SELECT
    ws.device_type,
    SUM(o.price_usd) AS total_revenue,
    COUNT(DISTINCT o.order_id) AS total_orders
FROM website_sessions ws
JOIN orders o
    ON ws.website_session_id = o.website_session_id
GROUP BY ws.device_type
ORDER BY total_revenue DESC;
```
